#### Обработка аномалии по количеству заказанного

```python
#Основное число покупок приходится на количество товара до определённого значения - 266.
# Оставим для исследования данные, где кол-во позиций до  266 шт.
#Уберем из данных все, где количество позиций в заказе больше 266 (75%). 
data_266 = data.query('quantity <= @q')
```

```python
#посмотрим общую информацию о данных в столбце
data_266['quantity'].describe()
```

```python
#рассмотрим график "количества товаров в заказе"
data_266.boxplot(column='quantity', figsize=(6,7))

plt.ylabel('количество товаров в заказе')
plt.title('Диаграмма размаха количества товаров в заказе max200')
plt.ylim(-20, 266)
plt.show()
```

```python
#Видим,что медиана,  75% совпали в 1, наше количество товаров в заказе заканчивается на 200, можем еще сократить наши данные
#обозначим количество товаров в заказе ревное 200 как q200
q200 = 200
```

```python
#Уберем из данных все, где количество позиций в заказе больше 200 
data_new = data.query('quantity <= @q200')
```

```python
#посмотрим общую информацию о данных в столбце
data_new['quantity'].describe()
```

```python
#напишем функцию для построения графиков сравнения между группами покупателей с 1 покупкой и больше 1 покупки
def compare(column1, column2, column3, title, y_label):  
    df1 = one_order_by_customers.reset_index().drop([column1, column2, column3], axis=1)
    df2 = two_orders_by_customers.reset_index().drop([column1, column2, column3], axis=1)
    df = pd.merge(df1, df2, on='prod_categ')

    df.plot.bar(x='prod_categ', figsize = (12,5))
    plt.xticks(rotation=50, fontsize=10)
    plt.xlabel("категория")
    plt.ylabel(y_label)
    

    plt.legend(['покупатели, совершившие 1 покупку', 'покупатели, совершившие больше 1 покупки'])
    plt.title(title)
    
    plt.show()
```

### RFM анализ. Сегментация покупателей.

```python
#Расчет давности Здесь мы рассчитываем давность для клиентов, совершивших покупку в магазине

df_recency = data_new.groupby(by='customer_id', 
                        as_index=False)['month'].max()
df_recency.columns = ['customer_id', 'last_purchase_date']
recent_date = df_recency['last_purchase_date'].max()
df_recency['recency'] = df_recency['last_purchase_date'].apply(
    lambda x: (recent_date - x).days)
df_recency.head()
df_recency.describe()
```

```python
#Расчет частоты. Здесь мы рассчитываем частоту покупок клиента 
frequency_df = data_new.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['month'].count()
frequency_df.columns = ['customer_id', 'frequency']
frequency_df.head()
frequency_df.describe()
```

```python
#Расчет денежной стоимости Здесь мы рассчитываем денежную оценку затрат клиентов на покупку товаров
#df['total'] = data_new.groupby(by = 'customer_id)['revenue'].sum()
monetary_df = data_new.groupby(by='customer_id', as_index=False)['revenue'].sum()
monetary_df.columns = ['customer_id', 'monetary']
monetary_df.head()
monetary_df.describe()
```

```python
#Объединение всех трех столбцов в одну таблицу данных

rf_df = df_recency.merge(frequency_df, on='customer_id')
rfm_df = rf_df.merge(monetary_df, on='customer_id').drop(
    columns='last_purchase_date')
rfm_df.head()
```

Далее для сегментации потребителей будем использовать квинтили — составление четырех равных частей на основе доступных значений — для расчета показателя RFM.

```python
quantiles = rfm_df.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
##  RFM Segmentation ----
RFM_Segment = rfm_df.copy()
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def R_Class(x,p,d):
    if x <= d[p][0.25]:  #хорошо, недавно покупали
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
# monetary and frequency Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FM_Class(x,p,d):
    if x <= d[p][0.25]:  #редко покупает и мало тратит
        return 1
    elif x <= d[p][0.50]: 
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
RFM_Segment['R_Quartile'] = RFM_Segment['recency'].apply(R_Class, args=('recency',quantiles,))
RFM_Segment['F_Quartile'] = RFM_Segment['frequency'].apply(FM_Class, args=('frequency',quantiles,))
RFM_Segment['M_Quartile'] = RFM_Segment['monetary'].apply(FM_Class, args=('monetary',quantiles,))
RFM_Segment['RFMClass'] = RFM_Segment.R_Quartile.map(str) \
                            + RFM_Segment.F_Quartile.map(str) \
                            + RFM_Segment.M_Quartile.map(str)
```

```python
RFM_Segment.head()
```

```python
#выделим классы по RFM
RFM_Segment.sort_values(by=['RFMClass','monetary'],ascending=[False, False])
```

```python
#посчитаем сумму значений RFM по ней будем проводить сегментацию
RFM_Segment['total']=RFM_Segment['R_Quartile']+RFM_Segment['F_Quartile']+RFM_Segment['M_Quartile']
RFM_Segment.describe()
```

```python
RFM_Segment.head(3)
```

```python
#функция для сегментации клиентов, будем делить по сумме баллов на 4 категории
def get_status(x):
    if x <= 3:  #хорошо, недавно покупали
        return 'потерянные клиенты'
    elif 3<x <=6:
        return 'спящие клиенты'
    elif 6<x <= 9: 
        return 'постоянные клиенты'
    else:
        return 'лучшие клиенты'
    
```

```python
#удалим лишние колонки
RFM_Segment['status']=RFM_Segment['total'].apply(get_status)

RFM_Segment.head(3)
```

```python
#посмотрим, как распределились клиенты,  по сколько их в группах

stat = RFM_Segment.groupby(by='status',as_index=False)['customer_id'].count()
stat.columns = ['status','val_count']

stat
```

```python
#посчитаем средний чек на покупателя
RFM_Segment['avg_bill'] = RFM_Segment['monetary']/RFM_Segment['frequency']
RFM_Segment.head(5)
```

```python
#посчитаем средний чек на покупателя
RFM_Segment['avg_bill'] = RFM_Segment['monetary']/RFM_Segment['frequency']
RFM_Segment.head(5)
```

```python
# посмотрим,какеи покупки делают сегментированные покупатели, найдем количество заказов и суммцу выручки
best_customers = data_all.groupby(['status', 'prod_categ'], as_index=False)\
    .agg({'order_id': 'nunique', 'revenue': 'sum'}).query('status == "лучшие клиенты"')

best_customers = round(best_customers.groupby('prod_categ')\
                            .agg({'revenue': 'sum', 'order_id': 'sum'}),2)
best_customers.columns = ['sum_revenue',  'sum_orders']
best_customers
```

```python
constant_customers = data_all.groupby(['status', 'prod_categ'], as_index=False)\
    .agg({'order_id': 'nunique', 'revenue': 'sum'}).query('status == "постоянные клиенты"')

constant_customers = round(constant_customers.groupby('prod_categ')\
                            .agg({'revenue': 'sum',  'order_id': 'sum'}),2)
constant_customers.columns = ['sum_revenue', 'sum_orders']
constant_customers
```

```python
#создадим таблицу изменения выручки по месяцам
data_const = data_all.pivot_table(
    index='month',
    columns='status',
    values='revenue',
    aggfunc='sum',
)
```

```python
sleeping_customers = data_all.groupby(['status', 'prod_categ'], as_index=False)\
    .agg({'order_id': 'nunique', 'revenue': 'sum'}).query('status == "спящие клиенты"')

sleeping_customers = round(sleeping_customers.groupby('prod_categ')\
                            .agg({'revenue': 'sum', 'order_id': 'sum'}),2)
sleeping_customers.columns = ['sum_revenue','sum_orders']
sleeping_customers
```

```python
lost_customers = data_all.groupby(['status', 'prod_categ'], as_index=False)\
    .agg({'order_id': 'nunique', 'revenue': 'sum'}).query('status == "потерянные клиенты"')

lost_customers = round(lost_customers.groupby('prod_categ')\
                            .agg({'revenue': 'sum', 'order_id': 'sum'}),2)
lost_customers.columns = ['sum_revenue',  'sum_orders']
lost_customers
```

```python
#напишем функцию для построения графиков сравнения выручки и суммы заказов по продукте в разрезе сегментированных клиентов 
def compare_all(column1, title1, y_label):  
    df1 = best_customers.reset_index().drop([column1], axis=1)
    df2 = constant_customers.reset_index().drop([column1], axis=1)
    df3 = sleeping_customers.reset_index().drop([column1], axis=1)
    df4 = lost_customers.reset_index().drop([column1], axis=1)
    df12 = pd.merge(df1, df2, on='prod_categ')
    df34 = pd.merge(df3, df4, on='prod_categ')
    df = pd.merge(df12, df34, on='prod_categ')

    df.plot.bar(x='prod_categ', figsize = (12,5))
    plt.xticks(rotation=50, fontsize=10)
    plt.xlabel("категория")
    plt.ylabel(y_label)
    

    plt.legend(['лучшие клиенты', 'постоянные клиенты','спящие клиенты', 'потерянные клиенты'])
    
    plt.title(title1)
    
    plt.show()
```

```python
compare_all( 'sum_orders',
        'Сравнение выручки сегментированных покупателей',
        'суммарная выручка')
```

Мы видим, что больше всего дохода от лучших , постоянных и "спящих" покупателей приходится на товары сад и город. Таже видим, что у постоянных покупателей пользуется спросом товары для дома.

## Проверка гипотез


Используем тест Крускала-Уоллиса для оценки первой гипотезы и проверки, разницы между средними чеками групп покупателей


```python
RFM_Segment.head(5)
```

**Гипотеза 1**

H0: Разницы между группами пользователей нет (средний чек, например)

H1: Между показателями наблюдаются существенные различия.

```python
#выполним тест Крускала-Уоллиса, используя функцию kruskal() из библиотеки scipy.stats:st.
#есть ли статистическая разница в среднем чеке сегментов покупателей
test = st.kruskal(RFM_Segment[RFM_Segment['status']=='постоянные клиенты']['avg_bill'],
RFM_Segment[RFM_Segment['status']=='спящие клиенты']['avg_bill'],
RFM_Segment[RFM_Segment['status']=='лучшие клиенты']['avg_bill'],
RFM_Segment[RFM_Segment['status']=='потерянные клиенты']['avg_bill'])

alpha=0.05

print("медианное значение среднего чека сегмента\"постоянные клиенты\":",
     RFM_Segment[RFM_Segment['status']=='постоянные клиенты']['avg_bill'].median())
print("медианное значение среднего чека сегмента\"спящие клиенты\":",
     RFM_Segment[RFM_Segment['status']=='спящие клиенты']['avg_bill'].median())
print("медианное значение среднего чека сегмента\"лучшие клиенты\":",
     RFM_Segment[RFM_Segment['status']=='лучшие клиенты']['avg_bill'].median())
print("медианное значение среднего чека сегмента\"потерянные клиенты\":",
     RFM_Segment[RFM_Segment['status']=='потерянные клиенты']['avg_bill'].median())

print('pvalue', test[1])
if test[1] < alpha:
    print('отвергаем нулевую гипотезу, разница есть') 
else:
    print('не можем отвергнуть нулевую гипотезу, разницы между сегментами нет')
```

pvalue=3.676920319405979e-56  меньше 0,05, мы можем отвергнуть нулевую гипотезу о том, что "Разницы между группами пользователей нет (средний чек, например)". Разница есть.
